```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
df=pd.read_csv('D:Fig_1/number_change.csv',encoding='gbk')

tm=list(df['time'])
num1=list(df['increase'])
num2=list(df['decrease'])
total_nodes_number=list(df['total'])
all_number=list(df['all'])
fig=plt.figure(figsize=(18,10))
ax1=fig.add_subplot(1,1,1)

#number of components
x=tm
y1=np.array(num1)
y2=np.array(num2)
y3=np.array(total_nodes_number)
y4=np.array(all_number)

ax1.plot(x,y4,c='red',marker='o',label='cumulative number of components',
         linestyle='--',linewidth=2.5)
ax1.plot(x,y3,c='red',marker='*',label='number of active components',
         linestyle='--',linewidth=2.5)

ax1.plot(x,y1,c='red',marker='x',label='number of new appearing components',linestyle='--',linewidth=2.5)
ax1.plot(x,y2,c='red',marker='^',label='number of disappeared components',linestyle=':',linewidth=2.5)

for a,b in zip(x,y3):
    plt.text(a, b+10, '%.0f' % b, ha='center', va= 'bottom',fontsize=8)
for a,b in zip(x,y1):
    plt.text(a, b, '%.0f' % b, ha='center', va= 'bottom',fontsize=8)
for a,b in zip(x,y2):
    plt.text(a, b-45, '%.0f' % b, ha='center', va= 'bottom',fontsize=8)
for a,b in zip(x,y4):
    plt.text(a, b+10, '%.0f' % b, ha='center', va= 'bottom',fontsize=8)

ax1.set_xlabel('time',fontsize=20)
ax1.set_ylabel('number of components',fontsize=20)
ax1.set_xticklabels(x, rotation=90,fontsize=16)
plt.tick_params(axis='y',colors='red')
ax1.set_yticklabels([0,0,500,1000,1500,2000,2500,3000], fontsize=16)


ax1.axvline(0,ls="--",linewidth=0.3,c="black")
ax1.axvline(1.2,ls="--",linewidth=0.3,c="black")
ax1.axvline(4.81,ls="--",linewidth=0.3,c="black")
ax1.axvline(8.6,ls="--",linewidth=0.3,c="black")
ax1.axvline(12.68,ls="--",linewidth=0.3,c="black")
ax1.axvline(15.44,ls="--",linewidth=0.3,c="black")
ax1.axvline(18.11,ls="--",linewidth=0.3,c="black")
ax1.axvline(19,ls="--",linewidth=0.3,c="black")


ax1.text(-0.10,2800,'before the\nend of Han\nDynasty'.center(1),fontsize=15)
ax1.text(2.1,2720,'Wei,Jin,\nSouthern and\nNorthern\nDynasties'.center(3),fontsize=15)
ax1.text(6.1,2800,'Sui,Tang\nand Five\nDynasties'.center(8),fontsize=15)
ax1.text(10,2800,'Song,Jin\nand Yuan\nDynasties'.center(11),fontsize=15)
ax1.text(13.8,2900,'Ming\nDynasty'.center(14),fontsize=15)
ax1.text(16,2900,'Qing\nDynasty'.center(16),fontsize=15)
ax1.text(18.1,2900,'Modern\nTimes',fontsize=15)


df1=pd.read_csv('D:Fig_1/formula_number.csv',encoding='gbk')

number=list(df1['formula_number'])
cumulative_number=list(df1['cumulative_number'])
tm=list(df1['time'])


#number of formulas
ax2=ax1.twinx()
#x_1=tm
y_1=np.array(number)
y_2=np.array(cumulative_number)
ax2.plot(x,y_2,c='blue',marker='^',label='cumulative number of TCM formulas',
         linestyle='--',linewidth=2.5)
ax2.plot(x,y_1,c='blue',marker='o',label='number of new TCM formulas in each period',
         linestyle=':',linewidth=2.5)



for a,b in zip(x,y_1):
    plt.text(a, b, '%.0f' % b, ha='center', va= 'bottom',fontsize=8)
for a,b in zip(x,y_2):
    plt.text(a, b, '%.0f' % b, ha='center', va= 'bottom',fontsize=8)


ax2.set_xlabel('time',fontsize=20)
ax2.set_ylabel('number of formulas',fontsize=20)
plt.tick_params(axis='y',colors='blue')
ax2.set_yticklabels([0,0,10000,20000,30000,40000,50000,60000], fontsize=16)


handles1, labels1=ax1.get_legend_handles_labels()
handles2, labels2=ax2.get_legend_handles_labels()
font1 = { 'size': 20 }
plt.legend(handles1+handles2,labels1+labels2,loc='lower center',
           bbox_to_anchor=(0.5, -0.36),ncol=2,prop=font1 )

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False

plt.savefig('D:Fig_1/Fig1.TIFF',dpi=300,bbox_inches='tight')
plt.show()
```
