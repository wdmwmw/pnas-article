```python
#Fig.S9
fig = plt.figure(figsize=(20, 30))
n=1

for i in range(2,len(files)):
    time=files[i].replace('.csv','')
    print(time)
    position=file_path+files[i]
    df=pd.read_csv(position,encoding='gbk')
    
    ax1 = fig.add_subplot(6,3,n)
    ax1.set_title(time,fontsize=12)
    degree=list(df['degree'])
    d=degree[0:degree.index(0)]
    x, y = ccdf(d, linear_bins=True)
    ax1.scatter(x, y, color='g', s=8, label='degree data')
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15) 
    plt.title(time,fontsize=15)
    fit = Fit(d, discrete=True,xmin=min(d))
    print(fit.distribution_compare('power_law', 'lognormal'))

    fit.power_law.plot_ccdf(ax=ax1, color='#CD3F2C', linestyle='--', label='Power law fit')
    fit.lognormal.plot_ccdf(ax=ax1, color='#FF8FA2', linestyle='--', label='Lognormal fit')
    
    if n==1 or n==4 or n==7 or n==10 or n==13 or n==16:
        ax1.set_ylabel(u"p(Kâ‰¥k)",fontsize=15)
    if n>15:
        ax1.set_xlabel("k",fontsize=15)
    
    handles, labels = ax1.get_legend_handles_labels()
    ax1.legend(handles, labels, loc=3,fontsize=15)
    
    print('alpha:',fit.power_law.alpha)
    print('sigma:',fit.power_law.sigma)
    print('D:',fit.power_law.D)
    n+=1
    print(n)
fig.savefig('D:data/paper_1/method_1/pnas_fig/FigS9.TIFF',dpi=300,bbox_inches='tight')
plt.show()
```
