```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
```

```python
fig=plt.figure(figsize=(10,8))
ls=['ucinet','rich_club_1','result_Rombach','result_Rossa']


df=pd.read_csv('D:Fig_S15/ucinet.csv',encoding='gbk')
df.set_index('herb1',inplace=True)
norm = Normalize(vmin = 0,vmax = 4294)

ax1=plt.subplot(221)
ax1.imshow(df,norm = norm,cmap = 'rainbow')

ax1.set_xlabel('core components',fontsize=15)
ax1.set_ylabel('core components',fontsize=15)
ax1.set_xticks([])
ax1.set_yticks([])

ax1.set_title('continuous model method',fontsize=15)
ax1.text(-4.5,1.3,'A',fontsize=20)


df=pd.read_csv('D:Fig_S15/rich_club.csv',encoding='gbk')
df.set_index('herb1',inplace=True)

ax2=plt.subplot(222)
ax2.imshow(df,norm = norm,cmap = 'rainbow')
ax2.set_xlabel('core components',fontsize=15)
ax2.set_ylabel('core components',fontsize=15)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_title('rich club method',fontsize=15)
ax2.text(-4.0,1,'B',fontsize=20)

df=pd.read_csv('D:Fig_S15/result_Rombach.csv',encoding='gbk')
df.set_index('herb1',inplace=True)

ax2=plt.subplot(223)
ax2.imshow(df,norm = norm,cmap = 'rainbow')
ax2.set_xlabel('core components',fontsize=15)
ax2.set_ylabel('core components',fontsize=15)
ax2.set_xticks([])
ax2.set_yticks([])
ax2.set_title('continuous model\nextension method',fontsize=15)
ax2.text(-4.0,1,'C',fontsize=20)

df=pd.read_csv('D:Fig_S15/result_Rossa.csv',encoding='gbk')
df.set_index('herb1',inplace=True)

ax3=plt.subplot(224)
ax3.imshow(df,norm = norm,cmap = 'rainbow')
ax3.set_xlabel('core components',fontsize=15)
ax3.set_ylabel('core components',fontsize=15)
ax3.set_xticks([])
ax3.set_yticks([])
ax3.set_title('random walk method',fontsize=15)
ax3.text(-3.2,0.8,'D',fontsize=20)


sc=ax1.imshow(df,norm = norm,cmap = 'rainbow')
cbar_ax=fig.add_axes([0.90,0.2,0.02,0.68])
fig.colorbar(sc, cax=cbar_ax)
plt.subplots_adjust(left=0.3, bottom=0.2, right=0.9, top=None, 
                    wspace=0.1, hspace=0.39)


plt.savefig('D:Fig_S15/FigS15.TIFF',dpi=300,bbox_inches='tight')
plt.show()
```
