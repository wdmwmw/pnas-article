```python
#stability of core
file_path='D:Fig_3/core_nodes/'
files=os.listdir(file_path)
files.sort(key=lambda x:int(x.split('/')[-1].split('.')[0]))

final_df=pd.read_csv('D:Fig_3/core_nodes/1901_2016.csv',encoding='gbk')
final_core=list(final_df['core'])
time=[]
common=[]
PCCS=[]
difference=[]
stability=[]
for file in files:
    time.append(file.replace('.csv',''))
    position=file_path+file
    df=pd.read_csv(position,encoding='gbk')
    core=list(df['core'])

    PCCS.append(len(core))
    com=[item for item in core if item in final_core]
    common.append(len(com))
    difference.append(1-len(com)/len(final_core))
    stability.append(len(com)/len(final_core))
new_df=pd.DataFrame()
new_df['time']=time
new_df['common']=common
new_df['PCCS']=PCCS
new_df['difference']=difference
new_df['stability']=stability
new_df.to_csv('D:Fig_3/stability_of_core.csv',index=False,encoding='gbk')
```
