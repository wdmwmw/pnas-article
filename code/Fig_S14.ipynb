```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(12,10))
ax4=plt.subplot(111)

df=pd.read_csv('D:Fig_S14/core/rich_club.csv',encoding='gbk')
time=list(df['time'])
y1=list(df['original'])
y2=list(df['common'])

df1=pd.read_csv('D:Fig_S14/core/result_Rombach.csv',encoding='gbk')
y3=list(df1['common'])

df2=pd.read_csv('D:Fig_S14/core/result_Rossa.csv',encoding='gbk')
y4=list(df2['common'])

ax4.plot(time,y1,c='Red',marker='x',linewidth=0.8,label='number of core components in Core')
ax4.plot(time,y2,c='green',marker='x',linewidth=0.8,
         label='numebr of common core components\nbetween Core and Core-A')
ax4.plot(time,y3,c='blue',marker='x',linewidth=0.8,
         label='numebr of common core components\nbetween Core and Core-B')
ax4.plot(time,y4,c='darkorange',marker='x',linewidth=0.8,
         label='numebr of common core components\nbetween Core and Core-C')

ax4.set_xlabel('time',fontsize=20)
ax4.set_ylabel('number of core components',fontsize=20)
ax4.set_xticklabels(time, rotation=90,fontsize=20)
ax4.set_ylim([-0.5,43])
ax4.set_yticklabels(['{:.0f}'.format(0),
                     '{:.0f}'.format(0),
                     '{:.0f}'.format(5),
                     '{:.0f}'.format(10),
                     '{:.0f}'.format(15),
                     '{:.0f}'.format(20),
                     '{:.0f}'.format(25),
                      '{:.0f}'.format(30),
                      '{:.0f}'.format(35),
                      '{:.0f}'.format(40)], fontsize=20)


df3=pd.read_csv('D:Fig_S14/stability/rich_club.csv',
               encoding='gbk')
y5=list(df3['stability'])
df4=pd.read_csv('D:Fig_S14/stability/result_Rombach.csv',
               encoding='gbk')
y6=list(df4['stability'])
df5=pd.read_csv('D:Fig_S14/stability/result_Rossa.csv',
               encoding='gbk')
y7=list(df5['stability'])
df6=pd.read_csv('D:Fig_S14/stability/ucinet.csv',
               encoding='gbk')
y8=list(df6['stability'])


ax10=ax4.twinx()
ax10.plot(time,y8,c='red',linestyle='--',marker='o',linewidth=0.8,
          label='stability of the Core')
ax10.plot(time,y5,c='green',linestyle='--',marker='o',linewidth=0.8,
          label='stability of the Core-A')
ax10.plot(time,y6,c='blue',linestyle='--',marker='o',linewidth=0.8,
          label='stability of the Core-B')
ax10.plot(time,y7,c='darkorange',linestyle='--',marker='o',linewidth=0.8,
          label='stability of the Core-C')


ax10.set_xlabel('time',fontsize=20)
ax10.set_ylabel('stability of the core',fontsize=20)
ax10.set_ylim([-0.05,1.15])
ax10.set_yticklabels(['{:.1f}'.format(0.0),
                     '{:.1f}'.format(0.0),
                     '{:.1f}'.format(0.2),
                     '{:.1f}'.format(0.4),
                     '{:.1f}'.format(0.6),
                     '{:.1f}'.format(0.8),
                     '{:.1f}'.format(1.0)], fontsize=20)
handles1, labels1=ax4.get_legend_handles_labels()
handles2, labels2=ax10.get_legend_handles_labels()



font1 = { 'size': 16 }
plt.legend(handles1+handles2,labels1+labels2,loc='lower center',
           bbox_to_anchor=(0.48, -0.56),ncol=2,prop=font1 )


#plt.savefig('D:data/paper_1/method_1/final_stage/total_difference/final/difference_stability_2016_0614.png',dpi=300,bbox_inches='tight')
plt.savefig('D:data/paper_1/method_1/pnas_fig/FigS14.TIFF',dpi=300,bbox_inches='tight')
plt.show()
```
