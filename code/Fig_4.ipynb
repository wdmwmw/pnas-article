```python
#Fig.4B
fig = plt.figure(figsize=(12,6))
ax1=fig.add_subplot(1,2,1)
simple_df=pd.read_csv('D:Fig_4/building_out.csv',encoding='gbk')
tuples=list(simple_df['tuple'])
com=list(simple_df['combination'])
no=list(simple_df['not'])
s=list(simple_df['sum'])
combinations=[com[i]/s[i] for i in range(len(com))]
non_combinations=[no[i]/s[i] for i in range(len(no))]
ax1.bar(tuples,combinations,label='build on smaller tuples',width=0.3,color='red',alpha=0.5)
ax1.bar(tuples,non_combinations,label='not build on smaller tuples',width=0.3,bottom=combinations,color='green',alpha=0.5)
ax1.set_xticklabels(tuples,fontsize=16)
ax1.set_ylim([-0.05,1.15])
ax1.set_yticklabels(['{:.1f}'.format(0),
                     '{:.1f}'.format(0),
                     '{:.1f}'.format(0.2),
                     '{:.1f}'.format(0.4),
                     '{:.1f}'.format(0.6),
                     '{:.1f}'.format(0.8),
                     '{:.1f}'.format(1.0)], fontsize=16)
def to_percent(tmp,position):
    return '%1.0f'%(100*tmp)+'%'
plt.gca().yaxis.set_major_formatter(FuncFormatter(to_percent))
ax1.set_ylabel('proportions of combinations',fontsize=18)
ax1.set_xlabel('type of tuples',fontsize=18)
ax1.legend(loc='lower center',bbox_to_anchor=(0.48, -0.24),ncol=2,fontsize=12)
ax1.text(-0.06,0.3,'build on 2-tuples',rotation=90,fontsize=12)
ax1.text(0.94,0.3,'build on 3-tuples',rotation=90,fontsize=12)
ax1.text(1.94,0.3,'build on 4-tuples',rotation=90,fontsize=12)
ax1.text(2.94,0.3,'build on 5-tuples',rotation=90,fontsize=12)
plt.savefig('D:Fig_4/2.TIFF',dpi=300,bbox_inches='tight')
plt.show()
```
